### Fazendo os imports dos pacotes necessarios

```python
# Libs para modelagem e Matrizes
import numpy as np
import pandas as pd

# Libs para analises graficas
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go

# Libs para ignorar avisos
import warnings

# Desabilitando avisos
warnings.filterwarnings('ignore')
```

### Trabalhando os dados

```python
# Lendo os dados
df_magalu = pd.read_excel('Vase_004 - Magalu - Sem Resolução.xlsx')

# Verificando
df_magalu.head()
```

```python
# Verificando o tamanho da planilha
df_magalu.shape
```

```python
# Verificando as informacoes
df_magalu.info()
```

```python
# Discriminando os dados da planilha
df_magalu.describe()
```

###Series Temporais

```python
# Colocando a data como index
dados = df_magalu.set_index('Data')

dados.head()
```

```python
# Visualizando os dados no grafico

plt.style.use('seaborn-darkgrid')
plt.figure( figsize=(16,5))
plt.title('Analise das ações da magalu - Fechamento', fontsize=14, loc='left')
plt.plot(dados.index, dados['Fechamento'])
plt.xlabel('Periodo da cotação')
plt.ylabel('Valor da Ação (R$)');
```

```python
# Verificando os ultimos dados da base de dados

dados.tail()
```

```python
# Analisando media movel de 5 em 5 dias
media_movel = dados['Fechamento'].rolling(5).mean()
media_tendencia = dados['Fechamento'].rolling(30).mean()
```

```python
# Analisando as medias

plt.style.use('seaborn-darkgrid')
plt.figure( figsize=(16,5))
plt.title('Analise das ações da magalu - Fechamento', fontsize=14, loc='left')
plt.plot(dados.index, dados['Fechamento'])
plt.plot(dados.index, media_movel)
plt.plot(dados.index, media_tendencia)
plt.xlabel('Periodo da cotação')
plt.ylabel('Valor da Ação (R$)');
```

```python
sns.boxplot( data=dados, x='Fechamento');
```

```python
# Boxplot Mensal

df_magalu['Mes'] = df_magalu['Data'].dt.month

df_magalu.head()
```

```python
plt.figure( figsize=(16,5))
sns.boxplot(data=df_magalu, x='Mes', y='Fechamento');
```

```python
# Visualização de forma mais analitica

df_magalu.groupby(['Mes']).describe()['Fechamento']
```

```python
# Grafico

grafico = go.Figure(
    data=[
          go.Candlestick(
              x = dados.index,
              open = dados['Abertura'],
              high = dados['Maior'],
              low = dados['Maior'],
              close = dados['Fechamento'],
          )

    ]
)

grafico.show()
```

```python
# Sem a barra inferior

grafico.update_layout(xaxis_rangeslider_visible=False)

grafico.show()
```
